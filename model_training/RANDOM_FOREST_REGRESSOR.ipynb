# SPLITTING DATA

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.42,random_state=40)
```

# RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
forest_regressor = RandomForestRegressor()
```

```python
forest_regressor.fit(X_train, y_train)
```

```python
print(" co-efficient of determination R^2 with train dataset : {}".format(forest_regressor.score(X_train, y_train)))
```

```python
print(" co-efficient of determination R^2 with test dataset : {}".format(forest_regressor.score(X_test, y_test)))
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(forest_regressor, X, y , cv=5)
```

```python
print(" The score of the model is : {}".format(score))
print("--"*50)
print(" The mean score of the model is : {}".format(score.mean()))
```

```python
prediction = forest_regressor.predict(X_test)
```

```python
sns.distplot(y_test-prediction)
```

# HYPER-PARAMETER TUNING

HERE WE ARE GOING TO USE RANDOMIZED SEARCH CV FOR OUR HYPERPARAMETER TUNING

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
'''
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
# Method of selecting samples for training each tree
# bootstrap = [True, False]
'''
```

```python
n_estimators = [int(x) for x in np.linspace(start=100, stop=1500, num=15 )]
criterion = ["squared_error", "mse", "absolute_error", "poisson"]
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
max_features = ["auto", "sqrt", "log2"]
min_samples_leaf = [1, 2, 5, 10]
min_samples_split = [2, 5, 10, 15, 100]
bootstrap = ["True","False"]
```

```python
params = {
    "n_estimators": n_estimators,
    "criterion" : criterion,
    "max_depth" : max_depth,
    "min_samples_leaf" : min_samples_leaf,
    "min_samples_split" : min_samples_split,
    "bootstrap" : bootstrap
}
```

```python
params
```

```python
rf_random = RandomizedSearchCV(forest_regressor, params,n_iter=100, scoring='neg_mean_squared_error', cv = 5, verbose=2, random_state=42, n_jobs=1)
```

```python
rf_random.fit(X_train,y_train)
```
